```python
# CLIP text encoder

from transformers import CLIPTextModel,CLIPTokenizer
import json

classes_dict = "../datasets/IndoorVG/VG-SGG-dicts.json"
with open(classes_dict, 'r') as f:
    classes = json.load(f)

print(classes.keys())
obj_classes = classes['label_to_idx']
pred_classes = classes['predicate_to_idx']


# CLIP Text Model
model_name="openai/clip-vit-base-patch16"

model = CLIPTextModel.from_pretrained(model_name)
tokenizer = CLIPTokenizer.from_pretrained(model_name)

embeddings = {}

# Encode text
for text in obj_classes.keys():
    inputs = tokenizer([text], return_tensors="pt", padding=True)

    outputs = model(**inputs)
    text_embeds = outputs.pooler_output

    embeddings[text] = text_embeds[0].tolist()

for text in pred_classes.keys():
    inputs = tokenizer([text], return_tensors="pt", padding=True)

    outputs = model(**inputs)
    text_embeds = outputs.pooler_output

    embeddings[text] = text_embeds[0].tolist()
```
